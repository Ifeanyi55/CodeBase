```python
from transformers import pipeline
```

```python
generator = pipeline('text-generation', model = 'gpt2')
generator("Hello, I am a data scientist")
```

```python
pipe = pipeline("text-generation",model = "gpt2")
pipe("Ifeanyi is a data scientist")
```

```python
from transformers import pipeline
classify = pipeline("text-classification")
classify("I love it here, because everybody seems nice")
classify("Dogs and cats are fun pets to have")
```

```python
from transformers import pipeline
image_classifier = pipeline("image-classification")
image_classifier("doggy.jpg")
```
